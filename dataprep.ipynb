**SPLIITING OUR DATA**

**NB** 

In splitting our data , we use **x** and **y** where **x** is to represent the input values or features 

And the **y** represents the target variable or the output value in our dataset.
In the code , **x** = df.drop('prognosis', axis=1) , we are creating **x** by dropping the 'prognosis' column (target value) from the original DataFrame df. This ensures that x contains all the other columns that serve as features or input values for your model.

For **y** = df['prognosis'], we assign the 'prognosis' column to **y**. This column contains the target variable values that correspond to the features in **x.** The model willbe able to  learn from the patterns in the features **x** and  predict the corresponding target variable values **y.**
